```python
#3. Use a 'for' loop to read csv's and compile into a df
path = dirIn
all_files = glob.glob('*.csv')
all_df = []

#make a df
for f in all_files:
    df = pd.read_csv(f, sep=',')
    df['MC'] = f.split('.')[-2]
    all_df.append(df)
merged_df = pd.concat(all_df, ignore_index=True, sort=True)
merged_df.head(2)
```

```python
#creates a new df with reflectance of 
csv_files = [f for f in os.listdir(dirIn) if f.endswith('.csv')]
mean_df = []

for entry in csv_files:
    global df
    df = pd.read_csv(entry)
    df.set_index('Wavelength',inplace=True)
    df['Mean_reflectance_'+entry] = ((df['measurement1']+df['measurement2']+df['measurement3']+df['measurement4']+df['measurement5'])/5)
    df = df['Mean_reflectance_'+entry]
    df.to_csv('mean_of_obs_'+entry,header =True)
    mean_df.append(df)
    #print(mean_df)
mean_df = pd.DataFrame(mean_df)
mean_df.head()
```

```python
index = mean_df.index
new_index = []
for item in index:
    item = item[17:-4]
    new_index.append(item)

mean_df.index = new_index
mean_df.head(2)
```
